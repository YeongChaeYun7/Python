# Matplotlib

프로그래밍 언어 및 수학적 확장 NumPy 라이브러리를 활용한 플로팅 라이브러리

그래프를 생성하고 변화를 나티내기 편리하다

머신러닝, 이미지 처리 등 다양한 분야에서 데이터 시각화 목적으로 많이 활용

*(개인적으로 참고를 위해 구글링을 통해 정리한 것이라 정확하지 않을 수 있다)*

```python
import numpy as np
import matplotlib.pyplot as plt # 관습적으로 plt로 표기
```

# 기본 그래프 생성과 입력

plot()함수로 선 (Line), 마커 (Marker) 그래프를 그릴 수 있다

인자로 y 값을 리스트로 입력하면 그래프가 출력된다

```python
# 리스트 1개를 인자로 입력
plt.plot([1, 2, 3, 4]) # y
# plot()은 입력 리스트의 값을 y값이라고 가정하고 x값을 자동으로 생성 
plt.show() # 그래프 출력
```

```python
# 리스트 2개를 인자로 입력
plt.plot([1, 2, 3, 4], [1, 4, 9, 16]) # 순서대로 x, y 값들로 인식
# plot()은 임의의 개수의 인자를 받을 수 있다
plt.show() 
```

파이썬 딕셔너리처럼 레이블이 있는 데이터 사용 가능

plot()인자로 레이블(키)을 입력해주고, data 파라미터에 딕셔너리를 지정

```python
data_dict = {'data_x': [1, 2, 3, 4, 5], 'data_y': [2, 3, 5, 10, 8]}

plt.plot('data_x', 'data_y', data=data_dict)
plt.show()
```

# 축 레이블 설정
xlabel(), ylabel()을 활용하여 그래프의 x,y축에 대한 레이블 표시 가능

labelpad 파라미터로 축레이블의 여백 지정 가능

loc 파라미터로 축레이블의 위치 지정 (Matplotlib 3.3 이후 버전만 가능)

  - xlabel() loc ({‘left’, ‘center’, ‘right’})

  - ylabel() loc ({‘bottom’, ‘center’, ‘top’})

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.xlabel('X-Label', labelpad=15) # 문자열로 레이블 입력
plt.ylabel('Y-Label', labelpad=30) # labelpad 축레이블의 padding 지정
plt.show()
```

# 범례(Legend) 표시

그래프에 데이터 종류를 표시하는 텍스트이다

matplotlib.pyplot애서 legend() 지원

```python
plt.plot([1, 2, 3, 4], [2, 3, 5, 10], label='Price ($)') # plot()에 label 문자열을 지정
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.legend() # 범례 호출만해도 적절한 위치로 자동 배치된다
plt.show()
```

labelpad 파라미터로 축 레이블의 Padding지정

legend() 함수의 loc 파라미터활용
(위의 축레이블은 x,ylabel()의 인자로 씀)

```python
# 범례의 위치 지정
plt.plot([1, 2, 3, 4], [2, 3, 5, 10], label='Price ($)')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.legend(loc=(1.0, 1.0)) # flot 주의
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [2, 3, 5, 10], label='Price ($)')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.legend(loc=(0.0, 0.0))# 'lower left'문자열로도 지정 가능
plt.show()
```

ncol()을 사용해 범례의 열 개수를 지정할 수도 있다

디폴트값은 1개

```python
plt.plot([1, 2, 3, 4], [2, 3, 5, 10], label='Price ($)')
plt.plot([1, 2, 3, 4], [3, 5, 9, 7], label='Demand (#)')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
# plt.legend(loc='best')          # ncol = 1
plt.legend(loc='best', ncol=2)    # ncol = 2

plt.show()
```

# 축 범위와 스케일 지정

축범위 지정

xlim(): X축이 표시되는 범위 지정/반환

ylim(): Y축이 표시되는 범위 지정/반환

axis(): X, Y축이 표시되는 범위 지정/반환

```python
plt.plot([1, 2, 3, 4], [2, 3, 5, 10])
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.xlim([0, 5])      # X축 범위. 각각 입력하거나 리스트, 튜플 형태로 받음
plt.ylim([0, 20])     # Y축 범위
#plt.axis([0, 5, 0, 20])  # X, Y축 범위 한 번에 지정 [xmin, xmax, ymin, ymax] 인자 4개 이하 error
plt.show()
# 입력값이 따로 없으면 데이터에 맞춰 자동 지정
```

axis()의 옵션

'on' / 'off' / 'equal' / 'scaled' / 'tight' / 'auto' / 'normal' / 'image' / 'square'

```python
# axis 옵션 지정은 문자열로 넣는다
plt.plot([1, 2, 3, 4], [2, 3, 5, 10])
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
# plt.axis('square') # 활용도 가장 높을 듯.  축 길이 동일하게 맞춤
plt.axis('scaled') # r길이 스케일에 맞춰 늘고 줄음
plt.show()
```

축 범위 반환

```python
plt.plot([1, 2, 3, 4], [2, 3, 5, 10])
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

x_range, y_range = plt.xlim(), plt.ylim() # X축, Y축의 범위를 각각 반환
print(x_range, y_range)

axis_range = plt.axis('scaled')#  X, Y축의 범위 한 번에 반환
print(axis_range)
plt.show() 
```

# 막대 그래프

범주가 있는 데이터 값을 직사각형의 막대로 표현  
matplotlib.pyplot에서 bar() 함수 사용

```python
# 막대 그래프
x = np.arange(3)

years = ['2018', '2019', '2020'] # x
values = [100, 400, 900] # y
colors = ['y', '#000000', 'C2'] # 색상을 리스트로 입력하여 막대 각가 색깔 구분

plt.bar(x, values, color=colors,width=0.5) #width 막대 폭 지정. 기본 0.8
plt.xticks(x, years)

plt.show()
```

# 수평 막대 그래프(Horizontal bar graph)

matplotlib.pyplot의 barh() 사용  

이외 부가적이 기능은 위의 막대 그래프와 동일

```python
# 수평막대 그래프
y = np.arange(3)

years = ['2018', '2019', '2020'] # y
values = [100, 400, 900] # x

plt.barh(y, values)
plt.yticks(y, years)

plt.show()
```

```python
# plot()의 solid_capstyle, dash_capstyle를 사용해서 선의 끝 모양 지정
plt.plot([1, 2, 3], [4, 4, 4], linestyle='solid', linewidth=10,
      solid_capstyle='butt', color='C0', label='solid+butt')
plt.plot([1, 2, 3], [3, 3, 3], linestyle='solid', linewidth=10,
      solid_capstyle='round', color='C0', label='solid+round')

plt.plot([1, 2, 3], [2, 2, 2], linestyle='dashed', linewidth=10,
      dash_capstyle='butt', color='C1', label='dashed+butt')
plt.plot([1, 2, 3], [1, 1, 1], linestyle='dashed', linewidth=10,
      dash_capstyle='round', color='C1', label='dashed+round')


plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.axis([0.8, 3.2, 0.5, 5.0])
plt.legend(loc='upper right', ncol=2)
plt.tight_layout()
plt.show()
```

### font

plot()의 fontdict로 폰트 스타일 설정(축레이블 적용 가능?)

인자 안에 fontdict = {}로 표기하며

family’, ‘color’, ‘weight’, ‘size’와 같은 속성을 사용


```python
font1 = {'family': 'serif',
         'color': 'b',
         'weight': 'bold',
         'size': 14
         }

font2 = {'family': 'fantasy',
         'color': 'deeppink',
         'weight': 'normal',
         'size': 'xx-large'
         }

plt.plot([1, 2, 3, 4], [2, 3, 5, 10])
plt.xlabel('X-Axis', labelpad=15, fontdict=font1)
plt.ylabel('Y-Axis', labelpad=20, fontdict=font2)
plt.show()
```

# 그리드(격자) 배경 설정  
그리드 설정을 통해  데이터의 위치를 더 명확하게 표현 가능  
일반적인 격자무늬도 가능하고 축을 설정하여 한가지 축만 표현 가능  
matplotlib.pyplot 모듈의 grid() 사용

```python
# 기본 그리드 설정 plt.grid(True)
x = np.arange(0, 2, 0.2)

plt.plot(x, x, 'bo')
plt.plot(x, x**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(x, x**3, color='springgreen', marker='^', markersize=9)
plt.grid(True) # False 입력하면 기본 바탕 출력
plt.show()
```

```python
# x축만 출력
x = np.arange(0, 2, 0.2)

plt.plot(x, x, 'bo')
plt.plot(x, x**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(x, x**3, color='forestgreen', marker='^', markersize=9)
plt.grid(True, axis='y') # y축만 출력하는 경우는 axis='x'
plt.show()
```

```python
# 그리드도 스타일 지정 가능
x = np.arange(0, 2, 0.2)

plt.plot(x, x, 'g')
plt.plot(x, x**2, color='k')
plt.plot(x, x**3, color='b')
plt.grid(True, axis='y', color='red', alpha=0.5, linestyle='--')
plt.show()
```

# 컬러맵

matplotlib.pyplot은 컬러맵 관련 다양한 함수 지원


```python
# 산점도에 컬러맵 적용
np.random.seed(0)
arr = np.random.standard_normal((8, 100))

plt.subplot(2, 2, 1)
# plt.scatter(arr[0], arr[1], c=arr[1], cmap='spring')
plt.scatter(arr[0], arr[1], c=arr[1])
plt.spring()
plt.title('spring')

plt.subplot(2, 2, 2)
plt.scatter(arr[2], arr[3], c=arr[3])
plt.summer()
plt.title('summer')

plt.subplot(2, 2, 3)
plt.scatter(arr[4], arr[5], c=arr[5])
plt.autumn()
plt.title('autumn')

plt.subplot(2, 2, 4)
plt.scatter(arr[6], arr[7], c=arr[7])
plt.winter()
plt.title('winter')

plt.tight_layout()
plt.show()
```

```python
# 컬러바 표시
np.random.seed(0)
arr = np.random.standard_normal((8, 100))

plt.subplot(2, 2, 1)
plt.scatter(arr[0], arr[1], c=arr[1])
plt.viridis()
plt.title('viridis')
plt.colorbar() # 그래프 영역에 컬러바 포함시킴

plt.subplot(2, 2, 2)
plt.scatter(arr[2], arr[3], c=arr[3])
plt.plasma()
plt.title('plasma')
plt.colorbar()

plt.subplot(2, 2, 3)
plt.scatter(arr[4], arr[5], c=arr[5])
plt.jet()
plt.title('jet')
plt.colorbar()

plt.subplot(2, 2, 4)
plt.scatter(arr[6], arr[7], c=arr[7])
plt.nipy_spectral()
plt.title('nipy_spectral')
plt.colorbar()

plt.tight_layout()
plt.show()
```
